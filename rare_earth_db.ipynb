

***CHECKING DATA TYPES***
As we want to make math operations all values of columns we want to use for those calculations must be numbers.
If they are not we must create a fucntion to cast them to floats

```python
df_re.dtypes
```

```python
#as many numerical values are object type we cannot make numerical calculations with them. 
#We define a fucntion to convert them into floats

def into_float (df,list_of_columns):
    for col in list_of_columns:
        if col != 'NaN':
            df[col].replace(to_replace=['-'],value='nan',inplace=True)
            df[col]= df[col].str.replace(',', '.').astype(float)
        else:
            col
    return df
```

```python
list_of_columns= ['Resource (*10^4 t, REO, Total)', 'Grade (wt. %)',
       'La2O3', 'Ce2O3', 'Pr6O11', 'Nd2O3', 'Sm2O3', 'Eu2O3', 'Gd2O3', 'Tb4O7',
       'Dy2O3', 'Ho2O3', 'Er2O3', 'Tm2O3', 'Yb2O3', 'Lu2O3', 'Y2O3','HREE percentage']
```

```python
into_float(df_re, list_of_columns)
```
